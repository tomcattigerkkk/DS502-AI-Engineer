```python
import mxnet as mx 
```

```python
# def conv_layer 
def conv_layer(input_layer, n_kernel, kernel_size, kernel_stride, pool_size, pool_stride, pad, stride):
#     input_layer = mx.sym.Variable('input') # 1, 3, 512, 512
    l = mx.sym.Convolution(data=input_layer, kernel=(3, 3), stride=(1, 1), pad=(1, 1), num_filter=32)

    l = mx.sym.BatchNorm(l)
    l = mx.sym.Activation(l, act_type='relu')
    l = mx.sym.Pooling(l, pool_type='max', kernel=(2, 2), stride=(2, 2))
    
    
def get_conv(n, input_data):
    l = conv_layer(n, )
```
